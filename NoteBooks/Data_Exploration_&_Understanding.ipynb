# Importing Libraries & load data

# Visualization

```python
# Check the distribution of the IncidentGrade variable (if it exists in the dataset)
if 'IncidentGrade' in df.columns:  
    print("IncidentGrade Variable Distribution:")
    sns.countplot(df['IncidentGrade'])
    plt.title("Distribution of IncidentGrade Variable")
    plt.show()
```

```python
# Check for class imbalance
if 'IncidentGrade' in df.columns:
    target_counts = df['IncidentGrade'].value_counts()
    plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title("Class Imbalance of IncidentGrade")
    plt.show()
```

```python
# Top 10 Categories for Alert Titles
plt.figure(figsize=(10, 6))
alert_title_counts = df['AlertTitle'].value_counts().nlargest(10)
sns.barplot(x=alert_title_counts.values, y=alert_title_counts.index, hue=alert_title_counts.index, palette='plasma', legend=False)
plt.title("Top 10 Alert Titles")
plt.xlabel("Count")
plt.ylabel("Alert Title")
plt.show()
```

```python
# Select only numerical columns
numerical_data = df.select_dtypes(include=[np.number])

# Calculate correlation matrix for numerical features
correlation_matrix = numerical_data.corr()

plt.figure(figsize=(20, 10))    # Set the size of the plot
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f") # Plot the correlation matrix
plt.title("Correlation Heatmap of Numerical Features")
plt.tight_layout()
plt.show()
```
